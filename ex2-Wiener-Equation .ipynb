## Solving Wiener equation numerically 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# n is the number of time steps
# N is the number of realizations
n=1000
N=1000

w=np.zeros((N,n))

# dt -> time increment
delta_t=0.001
sqrt_delta_t=np.sqrt(delta_t)

for j in range(N):
    # producing a gaussian random time series for each realization
    f=np.random.normal(0,1,size=n)
    
    for i in range(1,n):
        # computing w(t) using Euler method 
        # initial value for w is 0
        w[j,i]=w[j,i-1]+sqrt_delta_t*f[i-1]
```

```python
# ploting w versus t for each realization    
plt.figure(figsize=(12,5))
for i in range(N):
    
    plt.plot(w[i,:])
plt.title('n = '+str(n)+', N = '+str(N)+', dt = '+str(delta_t))
plt.show()
```

```python
# verification of the relation stated in the report 

s_2=np.zeros(n)

for i in range(n):
    s_2[i]=sum(w[:,i]**2)
s_2=s_2/N
log_s_2=np.log(s_2)

t=np.array(range(n))
log_t=np.log(t*delta_t)
```

```python
# generating a line with slope = 1 in order to compare it with our result
x=np.array(range(-7,1))
y=x
```

```python
plt.figure(figsize=(10,6))
plt.plot(log_t,log_s_2,color='r',label='computed data')
plt.plot(x,y,color='b',label='theoritical expectation: y=x')
plt.title('log(<W^2(t)>) versus log(t) ')
plt.legend()
plt.show()
```
